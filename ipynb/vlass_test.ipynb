```python
from astroquery.cadc import Cadc
from astropy import units as u
import matplotlib.pyplot as plt
from astroquery.simbad import Simbad
target_name = 'IRAS 00500+6713'
result_table = Simbad.query_object(target_name)

cadc = Cadc()
radius = 0.03*u.deg
result = cadc.get_images(target_name, collection="VLASS", radius = radius)
```

```python
import numpy as np
from astropy import wcs
image = result[0][0].data
w = wcs.WCS( result[0][0].header, naxis=2)
nt, nt, nx, ny = image.shape

ax = plt.subplot(111,projection=w)
plt.imshow(image[0][0])#, vmin = -5)
#plt.scatter(nx/2, ny/2, color="r", facecolor=None)
plt.colorbar()
#plt.scatter(result_table["RA"], result_table["DEC"], color="r", s =100)
plt.show()

image = result[1][0].data
w = wcs.WCS( result[0][0].header, naxis=2)
nt, nt, nx, ny = image.shape

ax = plt.subplot(111,projection=w)
plt.imshow(image[0][0])#, vmin = -5)
#plt.scatter(nx/2, ny/2, color="r", facecolor=None)
plt.colorbar()
#plt.scatter(result_table["RA"], result_table["DEC"], color="r", s =100)
```
